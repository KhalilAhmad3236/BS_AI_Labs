```python
#password strength checker

attempts = 5

for attempt in range(1, attempts + 1):
    password = input(f"Attempt {attempt}: Enter your password: ")

    score = 0

    # Condition 1: Length
    if len(password) >= 8:
        score += 1

    # Condition 2: Contains number
    if any(char.isdigit() for char in password):
        score += 1

    # Condition 3: Contains uppercase letter
    if any(char.isupper() for char in password):
        score += 1

    # Feedback
    if score == 0:
        strength = "Very Weak"
    elif score == 1:
        strength = "Weak"
    elif score == 2:
        strength = "Moderate"
    else:
        strength = "Strong"

    print(f"Password Strength: {strength}")

    remaining = attempts - attempt
    print(f"Remaining Attempts: {remaining}")

    # Winning condition
    if score == 3:
        print(f"You created a strong password!")
        print(f"AI Training Level: Beginner → Intermediate")
        break

# Losing condition
if score != 3:
    print(f"Final Result: Password is not strong enough.")
```
