```python
# TASK-01 Solution

# Step 1: Ask the user for their full name
full_name = input("Enter your full name: ")

# Step 2: Ask the user for their favorite color
favorite_color = input("Enter your favorite color: ")

# Step 3: Ask the user for their birth year (as integer)
birth_year = int(input("Enter your birth year (e.g., 2000): "))

# Step 4: Calculate the current age based on the year 2026
current_year = 2026
age = current_year - birth_year

# Step 5: Print the three nice lines using f-strings only
print(f"Welcome, {full_name}!")
print(f"Your favorite color is {favorite_color} – perfect for calm AI thinking.")
print(f"You were born in {birth_year} → you are currently {age} years old ({current_year}).")
```

```python
# TASK-02 Solution

# Step 1: Ask user for two numbers (float allows decimals)
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Step 2: Ask user for the operation as a string
operation = input("Enter operation (+, -, *, /, **, %): ")

# Step 3: Validate operator
if operation not in ["+", "-", "*", "/", "**", "%"]:
    print("Invalid operator!")  # If operator is not valid
else:
    # Step 4: Handle division by zero case
    if operation == "/" and num2 == 0:
        print("Cannot divide by zero!")
    else:
        # Step 5: Perform calculation based on operator
        if operation == "+":
            result = num1 + num2
            print(f"{num1} + {num2} = {result}")
        elif operation == "-":
            result = num1 - num2
            print(f"{num1} - {num2} = {result}")
        elif operation == "*":
            result = num1 * num2
            print(f"{num1} * {num2} = {result}")
        elif operation == "/":
            result = num1 / num2
            print(f"{num1} / {num2} = {result:.2f}")
        elif operation == "**":
            result = num1 ** num2
            print(f"{num1} ** {num2} = {result:.2f}")
        elif operation == "%":
            result = num1 % num2
            print(f"{num1} % {num2} = {result}")
```

```python
# TASK-03 Solution

# -------------------------------
# Challenge A – Even/Odd & Multiple Checker
# -------------------------------

# Step 1: Ask user for starting and ending numbers
start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))

# Step 2: Use while loop to go through each number in range
current = start
while current <= end:
    # Step 3: Skip numbers divisible by 7 using 'continue'
    if current % 7 == 0:
        current += 1
        continue

    # Step 4: Check conditions for even and multiple of 5
    if current % 2 == 0 and current % 5 == 0:
        print(f"{current} → Even & Multiple of 5!!")
    elif current % 2 == 0:
        print(f"{current} → Even")
    elif current % 5 == 0:
        print(f"{current} → Multiple of 5!")
    else:
        print(f"{current} → Odd")

    # Step 5: Move to next number
    current += 1



```

```python
# TASK-03 Solution
# Challenge B – Password Strength Checker
# -------------------------------

# Step 1: Ask user for a password
password = input("\nEnter a password: ")

# Step 2: Check password strength using nested if/elif
if len(password) < 6:
    print("Too short!")
else:
    # Check if password has at least one digit
    if not any(char.isdigit() for char in password):
        print("Add at least one number")
    else:
        # Check if password has at least one uppercase letter
        if not any(char.isupper() for char in password):
            print("Add at least one capital letter")
        else:
            print("Strong password – good choice!")
```
